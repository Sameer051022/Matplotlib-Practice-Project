## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (you have the whole figure)
fig.savefig("../Matplotlib Practise/sample-plot.png")
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize=(10, 5))

# Plot to each different  axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize=(10, 5))

# Plot to each different  axis
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which one to use? (pyplot vs matplotlib OO method?)
* When plotting something quickly ok to use pyplot method
* when plotting something more advanced use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind = 'scatter',
            x= 'age',
            y = 'chol',
            c = 'target');
```

```python
## OO method mixed with pyplot
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind = 'scatter',
            x= 'age',
            y = 'chol',
            c = 'target',
            ax = ax);
#ax.set_xlim([45, 100]);
```

```python
over_50.target.values
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# plot data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
# customize plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# add legend
ax.legend(*scatter.legend_elements(), title="Target");

# add horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--')
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)
# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

# customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
       ylabel="Cholestrol")

# customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# add legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");

# add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target");

# add horizontal line ax0
ax0.axhline(y=over_50["chol"].mean(),
          linestyle='--');

# add horizontal line ax1
ax1.axhline(y=over_50["thalach"].mean(),
          linestyle='--');

# add title to figure
fig.suptitle("Heart disease Analysis", fontsize=16, fontweight="bold")
```

```python
# set style
plt.style.use("seaborn-v0_8-whitegrid")
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# plot data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter") # change color scheme
# customize plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# add legend
ax.legend(*scatter.legend_elements(), title="Target");

# add horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--')
```

```python
# customizing x and y axis limitations
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)
# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

# add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")

# customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
       ylabel="Cholestrol")
ax0.set_xlim([50, 80])

# customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
# add legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");

# add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target");

# add horizontal line ax0
ax0.axhline(y=over_50["chol"].mean(),
          linestyle='--');

# add horizontal line ax1
ax1.axhline(y=over_50["thalach"].mean(),
          linestyle='--');

# add title to figure
fig.suptitle("Heart disease Analysis", fontsize=16, fontweight="bold");
```
